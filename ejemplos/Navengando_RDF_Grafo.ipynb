```python
import rdflib
g = rdflib.Graph()
g.parse('http://dbpedia.org/resource/Berlin')
```

```python
# Iterando sobre un grafo
for s, p, o in g:
    print(s, p, o)
    break
```

```python
print(len(g))
```

```python
#Checa si el triple existe
from rdflib import URIRef

if(URIRef('http://dbpedia.org/resource/Berlin'), None, None) in g:
    print(f'El triple existe')
else:
    print(f'El triple no existe')

if(URIRef('http://dbpedia.org/resource/Paris'), None, None) in g:
    print(f'El triple existe')
else:
    print(f'El triple no existe')
```

```python
# Dar valor a las propiedades
from rdflib.namespace import RDF, RDFS, OWL, FOAF
properties1 = set()
for s, p, o in g:
    properties1.add(p)
    
from pprint import pprint
pprint(properties1)
```

```python
len(properties1)
```

```python
properties2= set()
for p in g.predicates():
    properties2.add(p)

pprint(properties2)
```

```python
print(len(properties2))
```

```python
# Iterar sobre una lista de etiquetas
for s, o in g.subject_objects(RDFS.label):
    print(o.value, o.language, o.datatype ,type(o))
    
for s, o in g.subject_objects(OWL.sameAs):
    print(o)
```

```python
population = URIRef('http://dbpedia.org/ontology/populationTotal')
for o in g.objects(None, RDFS.label):
    if o.language == 'en':
        print(o)
```

```python
#COnseguir un valor del grafico
from rdflib import Namespace
DBO = Namespace('http://dbpedia.org/ontology/')
DBR = Namespace('http://dbpedia.org/resource/')

print(g.value(DBR['Berlin'], DBO['populationTotal'], None) )
```
